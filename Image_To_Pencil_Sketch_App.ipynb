```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def img_to_sktch(img_path, k):
  img = cv2.imread(img_path)
  grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  inverted_img = cv2.bitwise_not(grey_img)
  blurred_img = cv2.GaussianBlur(inverted_img, (k,k), 0)
  inverted_blurred_img = cv2.bitwise_not(blurred_img)
  sketched_img = cv2.divide(grey_img, inverted_blurred_img, scale= 256.0)
  cv2.imwrite('/content/my_img.png', sketched_img)
  cv2_imshow(sketched_img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python
img_to_sktch('/content/image.png', 3)
```
